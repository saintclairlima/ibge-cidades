## Criando representação tabular dos dados

Como cada atributo vindo da API é uma coleção de valor por ano, vamos representar cada combinação valor-ano em uma coluna separada. Como nem todos os atributos têm valores para os mesmos anos (idh, por exemplo, não tem valor em 2025, ao passo que prefeito tem), vamos criar rótulos para todos os atributos.

Por exemplo, o valor abaixo:

```python
'docentes_ens_medio': {
  '2020': 50,
  '2021': 45,
  '2022': 30
}
```

vai ser convertido em:

```python
'docentes_ens_medio_2020': 50,
'docentes_ens_medio_2021': 45,
'docentes_ens_medio_2022': 30,
```

### Formatando os dados

````python
# criando combinações de rotulos-ano
rotulos_anos = set([])
for _, cidade in dados_rotulados.items():
    for atributo, valor in cidade.items():
        if isinstance(valor, dict):
            for ano in valor:
                rotulos_anos.add((atributo,ano,))
````

````python
# criando um dicionário modelo para estruturar os dados
# é criada também uma lista de atributos para ordenar os atributos alfabeticamente
modelo = {'codigo': None, 'municipio': None, 'estado': None, 'categoria': None, 'longitude': None, 'latitude': None, 'altitude': None}
atributos = [k for k in modelo.keys()]
outros_atributos = []
for atributo, ano in rotulos_anos:
    modelo[f'{atributo}_{ano}'] = None
    outros_atributos.append(f'{atributo}_{ano}')
outros_atributos.sort()
atributos += outros_atributos
````

````python
# extraindo os dados (achatando-os)
dados_achatados = {}
for id, cidade in dados_rotulados.items():
    cidade_achatada = modelo.copy()
    for atributo, valor in cidade.items():
        if isinstance(valor, dict):
            for ano, valor_ano in valor.items():
                cidade_achatada[f'{atributo}_{ano}'] = valor_ano
                cidade_achatada[atributo] = None
        else:
            cidade_achatada[atributo] = valor
    dados_achatados[id] = cidade_achatada
````

````python
# ordenando os atributos
dados_achatados = {
    id: {
        k: cidade[k] for k in atributos
    } for id, cidade in dados_achatados.items()}
````

#### Salvando o dicionário em .json

## Criando representação somente com dados mais recentes de cada atributo

Vamos representar os dados, também, mantendo somente os dados mais recentes de cada atributo. Por exemplo, o valor abaixo:

```python
'docentes_ens_medio': {
  '2020': 50,
  '2021': 45,
  '2022': 30
}
```
vai ser convertido em:
```python
'docentes_ens_medio': 30
```

````python
dados_recentes = {}
for id, cidade in dados_rotulados.items():
    cidade_recente = {}
    for atributo, valor in cidade.items():
        if isinstance(valor, dict):
            ano_recente = max(valor.keys())
            cidade_recente[atributo] = valor[ano_recente]
        else:
            cidade_recente[atributo] = valor
    dados_recentes[id] = cidade_recente
````
